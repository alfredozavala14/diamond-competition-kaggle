```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, Normalizer
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
```

```python
df_train = pd.read_csv("../data/clean_train_01.csv", index_col = 0)
df_test = pd.read_csv("../data/clean_test_01.csv", index_col = 0) 
```

```python
df_train.head()
```

## Model

```python
# select predictor and results
X = df_train.drop("price", axis = 1)
y = df_train.price
```

```python
# create train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=10)
```

### LinearRegression with all variables

```python
# create pipe
pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("linear", LinearRegression())
])

# Standarization & model Training
pipe.fit(X_train, y_train)

# Predicting
y_pred_train = pipe.predict(X_train)
y_pred_test = pipe.predict(X_test)

# Metrics
mse_train = mean_squared_error(y_pred_train,y_train)
mse_test = mean_squared_error(y_pred_test,y_test)

mse_train, mse_test
```

### LinearRegression with just carats

```python
# select predictor and results
X = df_train.carat
y = df_train.price
```

```python
# create train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=10)
```

```python
# create pipe
pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("linear", LinearRegression())
])

# Standarization & model Training
pipe.fit(X_train.values.reshape(-1, 1), y_train)

# Predicting
y_pred_train = pipe.predict(X_train.values.reshape(-1, 1))
y_pred_test = pipe.predict(X_test.values.reshape(-1, 1))

# Metrics
mse_train = mean_squared_error(y_pred_train,y_train)
mse_test = mean_squared_error(y_pred_test,y_test)

mse_train, mse_test
```

Higher mse than first model

### Random Forest

```python
# select predictor and results
X = df_train.drop("price", axis = 1)
y = df_train.price
```

```python
# create train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=10)
```

```python
# random forest regressor model

# create forest
rf = RandomForestRegressor(n_estimators=100, max_depth=None, max_features='auto')

# model training
rf.fit(X_train, y_train)

# predictions
y_pred_train = rf.predict(X_train)
y_pred_test = rf.predict(X_test)

# Metrics
mse_train = mean_squared_error(y_pred_train,y_train)
mse_test = mean_squared_error(y_pred_test,y_test)

mse_train, mse_test
```

Now we can apply a grid search

```python
# define model and params

rf_model = RandomForestRegressor() 

params = {
    "n_estimators":[50, 100, 200, 600], 
    "max_depth":[4, 5, 6, None], 
    "min_samples_split":[2, 4, 8], 
    "min_samples_leaf":[2, 3, 4]
}
```

```python
# define score

mse = make_scorer(mean_squared_error)
```

```python
# run grid search

clf = GridSearchCV(estimator = rf_model, param_grid = params, scoring = mse, verbose=2)
```

```python
# fit model

clf.fit(X_train, y_train)
```

```python
# check best parameters

clf.best_params_
```

```python
# define best model

best = clf.best_estimator_
best
```

```python
# predict using best model

y_pred_train = best.predict(X_train)
y_pred_test = best.predict(X_test)
mse_train = mean_squared_error(y_pred_train,y_train)
mse_test = mean_squared_error(y_pred_test,y_test)
mse_train, mse_test
```

### Gradient Boosting

```python
# start model

gb = GradientBoostingRegressor(n_estimators=100, max_depth=None, max_features='auto')
```

```python
# fit model

gb.fit(X_train, y_train)
```

```python
# predict
y_pred_train = gb.predict(X_train)
y_pred_test = gb.predict(X_test)

# Metrics
mse_train = mean_squared_error(y_pred_train,y_train)
mse_test = mean_squared_error(y_pred_test,y_test)

mse_train, mse_test
```

### KNR

```python
# start model
knr = KNeighborsRegressor()

# fit model
knr.fit(X_train, y_train)

# predict
y_pred_train = knr.predict(X_train)
y_pred_test = knr.predict(X_test)

# Metrics
mse_train = mean_squared_error(y_pred_train,y_train)
mse_test = mean_squared_error(y_pred_test,y_test)

mse_train, mse_test
```

### AdaBoost Regressor

```python
# start model
ada = AdaBoostRegressor()

# fit model
ada.fit(X_train, y_train)

# predict
y_pred_train = ada.predict(X_train)
y_pred_test = ada.predict(X_test)

# Metrics
mse_train = mean_squared_error(y_pred_train,y_train)
mse_test = mean_squared_error(y_pred_test,y_test)

mse_train, mse_test
```
